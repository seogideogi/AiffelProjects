```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf # acf plot  pacf plot   
from statsmodels.tsa.arima_model import ARIMA # ARIMA 
import pmdarima as pm # Auto ARIMA 

import warnings

warnings.filterwarnings('ignore')
```

```python
ap = pd.read_csv('/aiffel/aiffel/time_series_basic/AirPassengers.csv')
ap.drop('Month', axis = 1, inplace = True)
```

```python
ap
```

```python
plt.plot(ap)
plt.show()
```

```python
# np.log  log transformation
ap_transformed = np.log(ap)
```

```python
ap_transformed
```

```python
plot_acf(ap_transformed) # acf plot,     
plt.plot();
```

```python
plot_pacf(ap_transformed) # pacf plot, lag 2   
plt.plot();
```

```python
ap_diff = ap_transformed.diff()
ap_diff = ap_diff.dropna()
```

```python
# lag 1 lag 12  lag 
plot_acf(ap_diff)
plt.plot();
```

```python
plot_pacf(ap_diff)
plt.plot();
```

acf pacf n = 12    .    .

ARIMA      .

```python
ap_transformed
```

```python
train_size = int(len(ap_transformed)*0.8)

ap_transformed_train = ap_transformed[:train_size]
ap_transformed_test = ap_transformed[train_size:]
```

 Auto ARIMA    

AutoARIMA   

* seasonal : False  non-seasonal model .
* suppress_warnings : ARIMA      .
* trace : True  (error)    .
* max_D : seasonal     .

```python
# AutoARIMA
# seasonal = True, max_D() = 12
model = pm.AutoARIMA(seasonal = True, suppress_warnings = True, trace = True, max_D = 12) 
res = model.fit(ap_transformed_train)
```

```python
# confidence interval 
preds, conf_int = res.predict(n_periods=ap_transformed_test.shape[0], return_conf_int=True) 

#      .
print(preds.shape) 
print(type(preds))
print('--'*40)
print(preds.shape[0])
print(type(preds.shape[0]))

print('--'*40)
print(preds)
print('--'*40)
print(conf_int)
```

```python
#   . 
# 0. train test data (axis) .
# 1. x test   train data , y train data  train data  .
# 2. 1  (AutoARIMA    )    .
# 3.     y(  ) (scatter plot) .
# 4. 2, 3      .
x_axis = np.arange(ap_transformed_train.shape[0] + preds.shape[0]) # 115 + 29, plot   x . # 0
plt.plot(x_axis[:ap_transformed_train.shape[0]],ap_transformed_train, alpha=0.75) # 1
plt.plot(x_axis[ap_transformed_train.shape[0]:], preds, alpha=0.75) # 2
plt.scatter(x_axis[ap_transformed_train.shape[0]:],ap_transformed_test, 
            alpha=0.4, marker='o')  # 3  test data scatter
plt.fill_between(x_axis[-preds.shape[0]:],
                 conf_int[:, 0], conf_int[:, 1],
                 alpha=0.1, color='b') # 4
plt.title("Log Transformed Air Passengers Forecast") # plot 
plt.show() # 4 plot  .
```

## ARCH  

```python
# -1 :     
import datetime as dt #      Python   .
import arch.data.sp500 # S&P 500  arch .

# -2 :      
st = dt.datetime(2002, 1, 1) # start date
en = dt.datetime(2022, 1, 1) # end date
data = arch.data.sp500.load() # s&p 500  
market = data["Adj Close"] #  
returns = 100 * market.pct_change().dropna() # return  #   !
ax = returns.plot() #   
xlim = ax.set_xlim(returns.index.min(), returns.index.max()) # xlim x    .   .
plt.show()
```

   ,  (Conditional Heteroskedasticity)  .

```python
# -3 : ARCH  
from arch import arch_model

am = arch_model(returns) # return arch 
res = am.fit(update_freq=5)

# -4 :    
print(res.summary())
#   ,   p-value, AIC, BIC  
#   volatility model  alpha[1], Beta[1] GARCH(1,1) , p-value   95%   
```
