상점과 구매자의 거리? ( 거리가 의미가 있을까. 여행 가면 ? 어차피 멀지 않나.) 

거리 구하겠음. 구매자 위도, 경도 일부 위치를 확인 해 보았을 때, 학습용 위치가 아닐까 싶음. 공터나 녹지 나옴.

```python
from geopy.distance import geodesic

# 거리 계산 함수 정의
def calculate_distance(row):
    merchant_coords = (row['merch_lat'], row['merch_long'])
    customer_coords = (row['lat'], row['long'])
    return geodesic(merchant_coords, customer_coords).kilometers

# 데이터프레임에 적용하여 새로운 열 추가
frd_df['distance_km'] = frd_df.apply(calculate_distance, axis=1)
```

시간대: trans_date_trans_time을 시간대로 분. 새벽, 아침, 오후, 저녁으로 나누면 특정 시간대에 사기가 발생할 확률을 파악 .

( 주로 새벽에 발생 활률이 높다고 함 )

```python
# trans_date_trans_time을 datetime 형식으로 변환
frd_df['trans_datetime'] = pd.to_datetime(frd_df['trans_date_trans_time'])
```

```python
# 시간대 나누기: 새벽 (0~6시)
def get_time_of_day(hour):
    if 0 <= hour < 6:
        return 'night'  # 가능 : 사기 시간대
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    else:
        return 'evening'
```

```python
frd_df['time_of_day'] = frd_df['trans_datetime'].dt.hour.apply(get_time_of_day)
```

```python
# 새벽에 발생한 거래 표시
frd_df['is_anomalous'] = frd_df['time_of_day'] == 'night'
```
