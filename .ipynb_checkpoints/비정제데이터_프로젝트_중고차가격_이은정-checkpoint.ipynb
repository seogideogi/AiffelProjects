```python
## 5.기존 데이터의 정보 70% 이상을 가지는 수준에서 최소한의 주성분 추출 (PCA)
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
threshold = 1e10  # 변경이 필요할 수 있음
numeric_df = numeric_df.applymap(lambda x: threshold if x > threshold else x)
```

```python
numeric_df = numeric_df.fillna(numeric_df.median()) # 중앙값 넣어주기
```

```python
numeric_df.info()
```

```python
numeric_df

numeric_df = numeric_df.drop(['title', 'Fuel type', 'Body type', 'Gearbox', 'Service history', 'country'],axis=1) ## 플롯형, object 제거
```

```python
pd.DataFrame(pca.fit_transform(numeric_df) , columns=['PC1', 'PC2'] )  # 얼마만큼의 정보 손실이 있는가 확인하면서 차원 찾아가기
```

```python
pca.explained_variance_ratio_.sum() # 대부분의 정보를 보유 하고 있다 0.9999~
```
